## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The outcome is observed 90-day CTR. Technically, the model will use `clicks_90d` as the response and `impressions_90d` as the number of opportunities for a click. CTR is then calculated as clicks divided by impressions.

This is an observed outcome from the starter dataset, not a product-defined label. However, it is measured in the same trailing 90-day window as the available features. Therefore, this analysis can describe conditional associations within this dataset, but it cannot claim to predict future CTR or prove causal effects.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

CSV_URL = (
    "https://raw.githubusercontent.com/"
    "flyrank-bih/flyrank-ml-internship-starter/main/"
    "data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(CSV_URL)

target_preview = df[
    ["content_id", "impressions_90d", "clicks_90d", "ctr"]
].copy()

target_preview["calculated_ctr_pct"] = (
    100 * target_preview["clicks_90d"]
    / target_preview["impressions_90d"]
)

display(target_preview.head())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can flag pages with a simple condition, such as high impressions and low CTR. However, it cannot reliably separate the relationship between CTR and several overlapping signals.

For example, average position strongly affects CTR, while content age, freshness, content type, search volume, and client-level differences may also be related to both position and CTR. A regression-style model can estimate the conditional association of each signal while accounting for the others, including nonlinear patterns and interactions where justified.

ML is not assumed to be better by default. If the expanded model does not improve on the position-only baseline or its effects are unstable across clients, a simpler descriptive analysis or rule is the more appropriate result.
