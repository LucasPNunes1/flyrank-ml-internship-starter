## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

CSV_URL = (
    "https://raw.githubusercontent.com/"
    "flyrank-bih/flyrank-ml-internship-starter/main/"
    "data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(CSV_URL)

summary = pd.DataFrame(
    {
        "metric": [
            "content items",
            "clients",
            "declining content items",
            "share of declining content items (%)",
        ],
        "value": [
            df["content_id"].nunique(),
            df["client_id"].nunique(),
            (df["trend_direction"] == "down").sum(),
            round((df["trend_direction"] == "down").mean() * 100, 1),
        ],
    }
)

display(summary)

content_type_summary = (
    df.groupby("content_type", dropna=False)
    .agg(
        pages=("content_id", "size"),
        pages_with_clicks=("clicks_90d", lambda values: (values > 0).sum()),
        total_impressions_90d=("impressions_90d", "sum"),
        total_clicks_90d=("clicks_90d", "sum"),
    )
    .assign(
        share_with_clicks_pct=lambda table: (
            100 * table["pages_with_clicks"] / table["pages"]
        ),
        weighted_ctr_pct=lambda table: (
            100 * table["total_clicks_90d"] / table["total_impressions_90d"]
        ),
    )
    .round(2)
    .sort_values("total_impressions_90d", ascending=False)
)

display(content_type_summary)
```
